### Exploring 'state' column 

```python
df['State'].unique()
```

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['State'],order=(df['State'].value_counts()).index,palette='Dark2')
plt.xticks(rotation=90)
```

```python
state_profit=df.groupby(['State'])['Profit'].sum()
state_profit.sort_values(ascending=False,inplace=True)
fig, ax = plt.subplots(figsize=(20,10))
plt.title("State vs Profit")
state_profit.plot.bar()
```

```python
profit_sales_per_state=df.groupby('State')['Profit','Sales'].agg(['sum'])
profit_sales_per_state.plot.bar(title="Total Profit and sales per state",figsize=(20,12))
plt.xticks(rotation=90)
plt.show(block=True)
```
